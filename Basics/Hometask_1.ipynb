```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

import scipy.stats as sts

plt.style.use('ggplot')
```

```python
lin = np.linspace(-1, 2, 100)
```

## 1.

### $F(x) = \frac{I_{[-1,2]}(x)(x+1)(x+3)}{15}$

### $p(x)=F^{'}(x)=\big(\frac{x^2+4x+3}{15}\big)_x^{'}=\frac{1}{15}(2x+4) \qquad \forall x\in[-1,2]$

### $\frac{1}{15}\int_{-1}^{2}(2x+4)dx=\frac{2}{15}\frac{x^2}{2}\bigg|_{-1}^{2}+\frac{4x}{15}\bigg|_{-1}^{2}=\frac{2}{15}(\frac{4}{2}-\frac{1}{2})+\frac{4}{15}(2+1)=\frac{1}{5}+\frac{4}{5}=1$

```python
Fx = lambda x: ((x+1)*(x+3))/15
px = lambda x: (2*x + 4)/15
```

```python
from scipy import integrate
```

```python
integrate.quad(lambda t: px(t), -1, 2)
```

```python
def stats(sample: np.array):
    print(f'Min: {sample.min()},\nMax: {sample.max()},\n\
diff: {np.abs(sample.min() - sample.max())}\nMean: {sample.mean()}\nMedian: {np.median(sample)}\nVar: {sample.var()}\n\
Var_off:{np.sum(sample - sample.mean())/(len(sample)-1)}\nVariation: {sts.variation(sample)}\nStd: {sample.std()}\nMedian: {np.median(sample)}\n\
Quantile-25%: {np.quantile(sample, 0.25)}\nQuantile-75%: {np.quantile(sample, 0.75)}\n\
Skew: {sts.skew(sample)}\nKurtosis: {sts.kurtosis(sample)}')
```

```python
stats(sample(1000))
```
